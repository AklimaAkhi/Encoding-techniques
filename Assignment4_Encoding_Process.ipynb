# LABEL ENCODER

```python
df2.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

# LabelEncoder with one by one column

```python
df2.sex= le.fit_transform(df2.sex)
df2.smoker= le.fit_transform(df2.smoker)
df2.region= le.fit_transform(df2.region)
```

```python
df2.head()
```

# LabelEncoder with loop

```python
df3.head()
```

```python
for column in df3.columns:
    if df3[column].dtype==np.number:
        continue
    else:
        df3[column]= le.fit_transform(df3[column])
```

```python
df3.head()
```

# To avoid warning

```python
from pandas.core.dtypes.common import is_numeric_dtype
for column in df4.columns:
    if is_numeric_dtype(df4[column]):
        continue
    else:
        df4[column]= le.fit_transform(df4[column])
```

```python
df4.head()
```

```python
cols = ['sex', 'smoker', 'region']
for column in cols:
    one = pd.get_dummies(df5[column], drop_first=True)
    df5= pd.concat([df5, one], axis=1).drop(column, axis=1)
#df5 = pd.get_dummies(df5, columns = cols)
```

# ORDINAL ENCODER

```python
df6.head()
```

```python
df6.sex.unique()
```

```python
gender=['female', 'male']
from sklearn.preprocessing import OrdinalEncoder
encoder= OrdinalEncoder(categories=[gender])
```

```python
encoded_data= encoder.fit_transform(df6[['sex']])
```

```python
encoded_data
```

```python
data= pd.DataFrame(encoded_data, columns=['sex'])
```

```python
data.head()
```

```python
df6=df6.drop('sex', axis=1)
```

```python
df6=pd.concat([df6, data], axis=1)
```

```python
df6.head()
```

```python
df7.head()
```

```python
cols = ['sex', 'smoker', 'region']
for col in cols:
    uniques= df7[col].unique()
    df7[col]= OrdinalEncoder(categories= [uniques]).fit_transform(df7[[col]])
```
